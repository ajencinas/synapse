# Download C4 Data
Downloads ~300 GB slice of allenai/c4 (English) into `data_c4/`.
Resumable — detects existing files and skips ahead.
Run `consolidate_data` -> `tokenizer_pipeline` -> `pre_train` after this.

```python
import os
from google.colab import drive
from datasets import load_dataset

drive.mount('/content/drive')

OUT_DIR = "/content/drive/MyDrive/synapse/datasets_pretrain/data_c4"
os.makedirs(OUT_DIR, exist_ok=True)

MAX_DOCS = 60_000_000
DOCS_PER_FILE = 10_000
SEPARATOR = "\n<|endoftext|>\n"

# Check what we already have (for resume)
existing_files = sorted([f for f in os.listdir(OUT_DIR) if f.startswith("c4_part_") and f.endswith(".txt")])
if existing_files:
    last_num = int(existing_files[-1].replace("c4_part_", "").replace(".txt", ""))
    docs_done = (last_num + 1) * DOCS_PER_FILE
    print(f"Found {len(existing_files)} existing files (~{docs_done:,} docs)")
    print(f"Resuming from doc {docs_done:,}...")
else:
    docs_done = 0

print(f"\nStreaming C4 English, target {MAX_DOCS:,} documents (~300 GB)...")
ds = load_dataset("allenai/c4", "en", split="train", streaming=True)

buf = []
file_idx = len(existing_files)
total_chars = 0

for i, example in enumerate(ds):
    if i >= MAX_DOCS:
        break
    # Skip already-done docs
    if i < docs_done:
        if i % 1_000_000 == 0 and i > 0:
            print(f"  Skipping... {i:,}/{docs_done:,}")
        continue

    buf.append(example["text"])
    if len(buf) >= DOCS_PER_FILE:
        path = os.path.join(OUT_DIR, f"c4_part_{file_idx:04d}.txt")
        content = SEPARATOR.join(buf)
        with open(path, "w", encoding="utf-8") as f:
            f.write(content)
        size_mb = os.path.getsize(path) / 1024 / 1024
        total_chars += len(content)
        total_gb = total_chars / 1024 / 1024 / 1024
        print(f"  {path.split('/')[-1]}: {size_mb:.1f} MB | {i+1:,} docs | {total_gb:.1f} GB total")
        buf = []
        file_idx += 1

# Write remainder
if buf:
    path = os.path.join(OUT_DIR, f"c4_part_{file_idx:04d}.txt")
    content = SEPARATOR.join(buf)
    with open(path, "w", encoding="utf-8") as f:
        f.write(content)
    total_chars += len(content)
    size_mb = os.path.getsize(path) / 1024 / 1024
    print(f"  {path.split('/')[-1]}: {size_mb:.1f} MB ({len(buf)} docs)")
    file_idx += 1

total_gb = total_chars / 1024 / 1024 / 1024
print(f"\nDone!")
print(f"  {i + 1:,} documents in {file_idx} files")
print(f"  Total: ~{total_gb:.1f} GB")
print(f"  Saved to: {OUT_DIR}")
print(f"\nNext: run consolidate_data -> tokenizer_pipeline (TRAIN_NEW_TOKENIZER=True) -> pre_train")
```
